```python
import numpy as np
#获取数据
from sklearn.model_selection import train_test_split
```

```python
# 注释：我已经把数据集改成简单的txt文件了，txt文件就是iris_data.txt，其中，Iris-setosa类用0表示，Iris-versicolor类用1表示，Iris-virginica类用2表示，“,”已经都替换为空格了
dataset = np.loadtxt('iris_data.txt')
# 打印dataset的大小，可以看出来dataset是一个150*5的矩阵
print(dataset.shape)
data = dataset[:,0:4]
target = dataset[:,4]
# data 是4维的特征
# target 是他们的种类
# 可以打印查看这两种数据
# print(data)
# print(target)
```

```python
X_train, X_test, y_train, y_test = train_test_split( data, target, random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
predict_result = knn.predict(X_test)
#对测试集进行预测
predict_result
```

```python
knn.score(X_test, y_test)
```
